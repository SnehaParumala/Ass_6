```python
import pandas as pd
import numpy as np
from sklearn import datasets
```

```python
data = pd.read_csv('/content/judge-1377884607_tweet_product_company (1).csv', encoding = 'Windows-1252')
```

```python
data.head()
```

```python
messages = data[['tweet_text','is_there_an_emotion_directed_at_a_brand_or_product']]
messages.columns = ['tweet','emotions']
```

```python
pd.set_option('display.max_colwidth', None)
messages.head()
```

```python
messages['emotions'] = messages['emotions'].map({"Negative emotion":0, "Positive emotion":1})
```

```python
messages.head()
```

```python
messages = messages.dropna()
```

```python
messages['emotions'] = messages['emotions'].astype(int)
```

```python
messages.head()
```

```python
#Tokenization
from keras.preprocessing import text
tokenizer = text.Tokenizer()
tokenizer.fit_on_texts(list(messages['tweet']))
tokenized_tweet = tokenizer.texts_to_sequences(messages['tweet'])
```

```python
from keras.utils import pad_sequences
x = pad_sequences(tokenized_tweet, maxlen = 100)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x, messages['emotions'].values, test_size = 0.2)
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Embedding, SimpleRNN, Dropout
```

```python
model = Sequential()
model.add(Embedding(input_dim = len(tokenizer.word_index)+1, output_dim = 128, input_length = 100))
model.add(SimpleRNN(10))
model.add(Dropout(0.5))
model.add(Dense(50,activation = 'relu'))
model.add(Dense(1,activation = 'sigmoid'))
```

```python
model.compile(loss = 'binary_crossentropy',
              optimizer ='adam',
              metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_train,y_train, epochs =10, validation_split = 0.1)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred[0]
```

```python
y_pred[y_pred>0.5] = 1
y_pred[y_pred<0.5] = 0
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
